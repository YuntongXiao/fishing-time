```python
#! pip install ***
# use pandas_datareader directly grab data from the Internet
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
%matplotlib inline
```

```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
```

```python
# Bank of America
BAC = data.DataReader("BAC", 'yahoo', start, end)
```

```python
# CitiGroup
C = data.DataReader("C", 'yahoo', start, end)
```

```python
# Goldman Sachs
GS = data.DataReader("GS", 'yahoo', start, end)
```

```python
# JPMorgan Chase
JPM = data.DataReader("JPM", 'yahoo', start, end)
```

```python
# Morgan Stanley
MS = data.DataReader("MS", 'yahoo', start, end)
```

```python
# Wells Fargo
WFC = data.DataReader("WFC", 'yahoo', start, end)
```

```python
tickers = ['t', 'i', 'c', 'k', 'e', 'r']
```

```python
stock_data=pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys='tickers')
#stock_data
```

```python
stock_data.columns.names = ['Bank Ticker','Stock Info']
```

```python
stock_data
```

```python
stock_data.xs(key='Close',axis=1,level='Stock Info').max()#xs is for multi index 
```

```python
returns=pd.DataFrame()
returns
```

```python
for t in tickers:
    returns[t+' Returns']=stock_data[t]['Close'].pct_change()
#returns.head()
returns[1:]
```

```python
sns.set_style('darkgrid')
sns.pairplot(data=returns[1:])
```
